```python
import numpy as np
from PIL import Image
import os
```

```python
path = '/home/dk/docrv2_sroie/DB/'
os.makedirs(path,exist_ok=True)
name = 'raw_test_task12'
description = 'data db from test_task12'
datadb = create_data_db(path=path, name=name, description=description, data_class='DATA')
```

```python
datadir = './test_task12_img/'
datadir_text = './test_task12_text/'
datadirinfo = glob(datadir+'*.jpg')
ind = 0
for idx in tqdm(range(len(datadirinfo))):
    try:
        imgfile = datadirinfo[idx]
        refname = os.path.basename((datadirinfo[idx]))[:-4]
        txtfile = datadir_text+refname+'.txt'
        img = Image.open(imgfile)

        f = open(txtfile,'r')
        txt = f.read().split('\n')
        char_table = pd.DataFrame({'x1':[], 'y1':[], 'x2':[], 'y2':[], 'x3':[], 'y3':[],
                                             'x4':[], 'y4':[], 'char':[]})
        char_table.x1 = char_table.x1.astype(int)
        char_table.y1 = char_table.y1.astype(int)
        char_table.x2 = char_table.x2.astype(int)
        char_table.y2 = char_table.y2.astype(int)
        char_table.x3 = char_table.x3.astype(int)
        char_table.y3 = char_table.y3.astype(int)
        char_table.x4 = char_table.x4.astype(int)
        char_table.y4 = char_table.y4.astype(int)
        chars = []
        for idx in range(len(txt)):
            if len(txt[idx])>0:
                ps = parse('{},{},{},{},{},{},{},{},{}',txt[idx])
                lps = list(ps)

                # save data
                chars.append(lps[8])
                char_data = pd.DataFrame({'x1':[int(lps[0])], 'y1':[int(lps[1])], 
                                             'x2':[int(lps[2])], 'y2':[int(lps[3])],
                                             'x3':[int(lps[4])], 'y3':[int(lps[5])],
                                             'x4':[int(lps[6])], 'y4':[int(lps[7])], 'char':lps[8]})
                char_table = char_table.append(char_data, ignore_index=True)
        #print(refname)
        #display(char_table)
        #img
        update_data(datadb, index = ind, label = 'task12', image = img, 
                    text = None, ref = refname, char_c = char_table, char = chars)
        ind = ind + 1
    except:
        print('fileerror : ',refname)
print_env(datadb)
```
